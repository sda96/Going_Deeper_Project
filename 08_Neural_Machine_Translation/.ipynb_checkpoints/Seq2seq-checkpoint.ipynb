#### 4.4 영어, 한국어 텍스트 전처리

```python
mecab = Mecab()
def ko_pre(sentence):
    return mecab.morphs(sentence)


def en_pre(sentence):
    return ["<sos>"] + sentence.lower().split(" ") + ["<eos>"]
```

```python
train["encoder_input"] = train["ko"].apply(lambda x : ko_pre(x))
train["pre_en"] = train["en"].apply(lambda x : en_pre(x))
train.head()
```

### 7. 인코더-디코더 모델 구축

```python
class Encoder(tf.keras.Model):
    
    def __init__(self, vocab_size, embedd_size, hidden_size):
        super(Encoder, self).__init__()
        self.embedd = tf.keras.layers.Embedding(input_dim = vocab_size,
                                                output_dim = embedd_size)
        self.lstm1 = tf.keras.layers.LSTM(hidden_size, return_sequences = True)
        self.lstm2 = tf.keras.layers.LSTM(hidden_size, return_sequences = True)
        self.lstm3 = tf.keras.layers.LSTM(hidden_size, return_sequences = True)
        self.lstm4 = tf.keras.layers.LSTM(hidden_size, return_sequences = True)
    
    def call(self, x):
        # shape(1,125,100)
        x = self.embedd(x)
        # shape(1,125,64)
        x = self.lstm1(x)
        x = self.lstm2(x)
        x = self.lstm3(x)
        x = self.lstm4(x)
        return x
```

```python
ex = reverse_encoder_input[0].reshape(1,-1)

encoder = Encoder(vocab_size, embedd_size, hidden_size)
context_vector = encoder.call(ex)
context_vector
```

```python
class Decoder(tf.keras.Model):
    
    def __init__(self, vocab_size, embedd_size, hidden_size):
        super(Decoder, self).__init__()
        self.embedd = tf.keras.layers.Embedding(input_dim = vocab_size,
                                                output_dim = embedd_size)
        self.lstm1 = tf.keras.layers.LSTM(hidden_size, return_sequences = True)
        self.outputs = tf.keras.layers.Dense(vocab_size, activation = "softmax")
        
    def call(self, x, context_vector):
        x = self.embedd(x)
        x = self.lstm1(x)
        x = tf.keras.layers.concatenate([x, context_vector])
        x = self.outputs(x)
        return x
```

```python
class Encoder_Decoder(tf.keras.Model):
    
    
    def __init__(self, vocab_size, embedd_size, hidden_size):
        super(Encoder_Decoder, self).__init__()
        self.vocab_size = vocab_size
        self.embedd_size = embedd_size
        self.hidden_size = hidden_size
        
        
        encoder = Encoder(self.vocab_size, self.embedd_size, self.hidden_size)
        decoder = Decoder(self.vocab_size, self.embedd_size, self.hidden_size)
        
    def call(self, x):
        encoder_input, decoder_input = x
        context_vector = encoder.call(encoder_input)
        outputs = decoder.call(decoder_input, context_vector)
        outputs = tf.argmax(outputs, axis = 2)
        outputs = tf.cast(outputs, tf.int32)
        return outputs
```

```python
encoder_decoder = Encoder_Decoder(vocab_size, embedd_size, hidden_size)
```
