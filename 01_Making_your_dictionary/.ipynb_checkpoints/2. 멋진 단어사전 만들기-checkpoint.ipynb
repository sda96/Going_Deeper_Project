## 1. 서론

오늘날 자연어를 의미가 담긴 숫자로 변환시켜서 기계어로 바꾸려는 다양한 노력을 하고 있습니다. 벡터화를 시키는 방법으로는 형태소 기반 벡터화와 예측 기반 벡터화가 존재합니다.  
형태소 기반 벡터화의 경우 한국어 형태소 분석기인 Mecab을 사용하여 문장을 형태소 단위로 분절하고 단어들의 빈도순으로 인덱스를 매핑시킨 뒤 기존의 문장들을 인덱스값으로 변환시키는 방식입니다.  
예측 기반 벡터화의 경우 OOV 문제와 UNK 문제를 해결하기 위해서 SetencePiece 모델을 사용할 것이며 해당 모델에는 uni-gram 방식과 byte-pair-encoding 방식이 존재하기에 두 방식 모두 실험하여 보겠습니다.

두 모델의 성능을 전처리 여부, 사전학습 여부, 단어장의 크기에 따라서 정확도(accuracy)를 사용하여 측정해보겠습니다.

정확도를 도출하기 위해서 사용된 데이터셋은 네이버 영화 댓글 감성 분석 데이터셋으로 약 20만개의 훈련 데이터와 5만개의 테스트 데이터로 구성되어 있으며 영화 댓글을 의미하는 documnet 칼럼과 해당 댓글의 긍부정 여부를 의미하는 label 칼럼이 존재합니다.  
- label이 0이면 부정, 1이면 긍정

### 1.1 성능 비교를 위한 모델들의 경우의 수

1. 전처리 전 형태소 분석기 (Mecab)
2. 전처리 후 형태소 분석기 (Mecab)
3. 전처리 후 형태소 분석기 + 사전 학습된 임베딩 벡터 (전이학습)
4. 전처리 전 SentencePiece (uni-gram)
5. 전처리 전 SentencePiece (byte-pair-enocding)
6. 전처리 후 SentencePiece (uni-gram)
7. 전처리 후 SentencePiece (byte-pair-enocding)
8. 형태소 분석기 (Mecab) + SetencePiece (uni-gram)
9. 형태소 분석기 (Mecab) + SetencePiece (byte-pair-enocding)

### 1.2 초기 환경 설정

batch_size = 128  
epochs = 5  
embedding_size = 200  
형태소 분석기 = Mecab()  
분류모델 = BiLSTM 이진 분류 모델  
hidden_size = 32

### 2.1 프로젝트 파이프라인의 parameters

#### add
- add = False 인 경우 기본 전처리
    - 중복되는 데이터 제거
    - 결측치 데이터 제거
- add = True 인 경우 추가 전처리
    - 텍스트 특수문자 전처리
    - 이상치 데이터 제거
    - 결측치 데이터 제거


#### vectorization_type
- vectorization_type = 0 인 경우 Mecab을 사용한 형태소 분석을 진행하고 빈도 기반 벡터화를 적용합니다.

- vectorization_type = 2 인 경우 Setencepiece 모델를 사용하며 byte-pair-encoding 방식을 적용합니다.  
- vectorization_type = 3 인 경우 Setencepiece 모델과 Mecab을 사용하며 uni-gram 방식을 적용합니다.  
- vectorization_type = 4 인 경우 Setencepiece 모델과 Mecab을 사용하며 byte-pair-encoding 방식을 적용합니다.

## 결론

각 모델들을 vocab_size와 추가 전처리 여부에 따라서 테스트셋에 대한 정확도(accuracy)를 구하였습니다.

|모델|추가 전처리 여부|vocab_size = 5,000|vocab_size = 10,000|
|-|-|-|-|
|Mecab|No|**86.05%**|**85.12%**|
|Mecab|Yes|84.77%|84.41%|
|Mecab + word2vec|Yes|84.41%|84.28%|
|spm unigram|No|85.46%|83.3%|
|spm bpe|No|85.26%|84.7%|
|spm unigram|Yes|83.58%|83.6%|
|spm bpe|Yes|84.06%|84.09%|
|Mecab + spm unigram|No|**79.86%**|80.32%|
|Mecab + spm bpe|No|80.17%|**78.9%**|

- vocab_size가 5,000 인 경우 전처리를 하기전의 Mecab모델이 약 86.05%로 가장 높았고 전처리전 Mecab + spm unigram모델이 약 79.86%로 가장 성능이 좋지 않았습니다.

- 전체적으로 전처리를 하기전의 모델들이 더 좋은 성능을 가지고 나온것으로 보아 현재 적용된 추가 전처리 단계에서 의미있는 정보까지 제외시켜버리는 문제가 발생하고 있다고 생각합니다. 혹은 전처리를 거치고 나면 학습하는 데이터의 수가 절반가량으로 감소하기 때문일지도 모르겠습니다.

- 인터넷을 찾아보며 Mecab과 spm모델을 동시에 사용하는 경우가 [가장 성능이 좋다는 글](https://monologg.kr/2020/05/02/koelectra-part1/)을 보았는데 실험설계가 잘못되었는지 가장 좋지 못한 성능을 내게 되었습니다.
    - 두 가지를 섞는 방식의 경우 코퍼스들을 Mecab으로 형태소 단위로 나눈 뒤 ```" ".join(list)```함수로 다시 문장들로 합쳐주었으며 형태소 단위로 공백이 있는 문장들을 다시 spm 모델에 적용하는 방식으로 진행하였습니다.
    
- 현재의 실험결과에 따르면 아무런 수정을 가하지 않은 순수한 형태의 형태소 단위의 벡터화가 가장 성능이 좋게 나온 것이 상당히 의외라고 생각합니다. 왜냐하면 적어도 spm모델의 bpe방식이 가장 성능이 좋을거라고 생각했기 때문입니다.

- 해당 실험을 통하여 아직까지 형태소 단위의 벡터화 방법의 성능이 생각보다 좋다는 것을 보았습니다.
